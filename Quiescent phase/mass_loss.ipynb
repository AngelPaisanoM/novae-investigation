## Mass transfer mechanism evaluations
We go first with the Bondy-Hoyle-Littleton mechanism that is written:
$$\dot M_{BHL}= \frac{1}{2} (\frac{GM_{WD}}{av_{\infty}^2})^2\dot M_{RG}$$

```python
#Evaluation of M_BHL 
# T CrB
G= 6.67*1E-11 # en mks
P= 227*86400 #orbital period (sec)
R_sun= 6.957E8  # en m


q1= 0.3198     #parallax as free param
Msum1= 1.729*1.989E30 # en kg

q2= 0.4927     #with gaia parallax
Msum2= 2.06*1.989E30  # en kg

a1 = ((P**2 * G * Msum1) / (4 * (np.pi**2)))**(1/3) # en m
a2 = ((P**2 * G * Msum2) / (4 * (np.pi**2)))**(1/3) # en m
a= (a1+a2)/2     #vamos a tomar el valor promdio

M_sun= 1.989*10**30 # en kg
Mwd=1.31*M_sun # en kg      (lower value)
v_wind= 3E4 #en m/s      # ~30 km/h
Mass_loss= (6.7*1E-8*M_sun)/(365.25*24*3600)    #considerando M_sc como valor referencial
M_BHL= 0.25*(((2*G*Mwd)/(a*v_wind**2))**2)*Mass_loss # en kg/s
M_BHL= M_BHL/(M_sun/ (365.25*24*3600))    #masas solares por año




print(M_BHL)
```

But it doesn't make sense to have an accretion rate higher than the mass loss rate, so we are missing something in the denominator. The solution would be to take the relative velocity of the wind which considers the orbital velocity (because the WD is also moving) so we retriave a lower value of $M_{BHL}$, let's see:

```python
import numpy as np

# ... (Previous constants G, P, Msum, a, etc. remain the same) ...

# 1. Calculate Orbital Velocity
# v_orb = 2 * pi * a / P
v_orb = (2 * np.pi * a) / P  # roughly 40,000 m/s for T CrB

# 2. Calculate Relative Velocity
# The wind moves radially, the star moves tangentially.
v_rel = np.sqrt(v_wind**2 + v_orb**2)

# 3. Update the BHL Formula
# Use v_rel instead of v_wind
M_BHL = 0.25 * (((2 * G * Mwd) / (a * v_rel**2))**2) * Mass_loss

# Check the fraction to ensure it makes sense (must be < 1)
accretion_efficiency = M_BHL / Mass_loss
print(f"Efficiency: {accretion_efficiency:.4f}")

M_BHL_solar = M_BHL / (M_sun / (365.25 * 24 * 3600))
print(f"Accretion Rate: {M_BHL_solar} M_sun/yr")
```

```python
#RS Oph (with the vrel correction included)

# Constants
G = 6.67e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30    # kg
R_sun = 6.957e8     # m
AU = 1.496e11       # m

# --- RS Oph Data ---
P_days = 453.6
P = P_days * 86400  # orbital period (sec)

# Stellar Parameters 
R_rg = ((50 + 80) * R_sun) / 2
M_wd = 1.35 * M_sun  # Hachisu et al 2007
M_rg = ((0.68 + 0.8) * M_sun) / 2  # Brandi et al average

# Orbital Parameters
q = M_rg / M_wd
a = ((1.425 + 1.502) * AU) / 2  # Average separation in meters

# Roche Lobe (Eggleton formula) 
R_rl = (a * 0.49 * q**(2/3)) / (0.6 * q**(2/3) + np.log(1 + q**(1/3)))

# Wind and Mass Loss
# Note: Mass_loss variable is the companion's mass loss rate
Mass_loss = (1.19e-7 * M_sun) / (365.25 * 24 * 3600)  # kg/s (from S-C formula)
v_wind = 1.5e4  # 15 km/s 

# --- THE CORRECTION ---

# 1. Calculate Orbital Velocity of the WD
#    v = 2 * pi * a / P
#    Note: 'a' is the separation. The relative velocity of the WD through the wind
#    is determined by the full orbital motion at separation 'a'.
v_orb = (2 * np.pi * a) / P

# 2. Calculate Relative Velocity (Wind + Orbit)
#    Wind moves radially, WD moves tangentially -> Pythagorean sum
v_rel = np.sqrt(v_wind**2 + v_orb**2)

print(f"Wind Velocity: {v_wind/1000:.1f} km/s")
print(f"Orbital Velocity: {v_orb/1000:.1f} km/s")
print(f"Relative Velocity (v_rel): {v_rel/1000:.1f} km/s")

# 3. Calculate BHL Accretion Rate using v_rel
M_BHL = 0.25 * (((2 * G * M_wd) / (a * v_rel**2))**2) * Mass_loss  # kg/s

# 4. Convert back to Solar Masses per Year
M_BHL_solar = M_BHL / (M_sun / (365.25 * 24 * 3600))

# 5. Check Efficiency
efficiency = M_BHL / Mass_loss

print(f"--------------------------------")
print(f"Mass Loss Rate (Donor): {1.19e-7} M_sun/yr")
print(f"Accretion Rate (BHL):   {M_BHL_solar:.3e} M_sun/yr")
print(f"Accretion Efficiency:   {efficiency:.2%}")
```


### RLOF
Now we go with RLOF considering the photospheric density values of $\rho \sim 8*10^{-6} kg/m^3$  from [Jan E.Staff+2016](https://ui.adsabs.harvard.edu/abs/2016MNRAS.458..832S/abstract) or $\rho \sim 9.3*10^{-7} kg/m^3$ for AGB stars [Noam Soker and Amos Harpaz+2002](https://ui.adsabs.harvard.edu/abs/2002astro.ph.10586S/abstract)

And computing it as:

$$    \dot M_0= \frac{2\pi}{\sqrt{e}}\left(\frac{\mathscr{R}T}{\mu} \right)^{3/2}\frac{R_{rl}^3}{GM_2}\rho_{ph}\,F(q)$$
so 
$$ \dot M_{R}=\dot M_0 \, exp \left(\frac{R_*-R_{rl}}{H_p} \right)$$

```python
#T CrB
# --- 1. Constants (SI Units) ---
G = 6.67e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30    # kg
R_sun = 6.957e8     # m
R_gas = 8314.0      # Ideal Gas Constant J/(kmol K) -> 8314 J/(kg K) if using molar mass in kg
mu = 1.2            # Mean molecular weight (neutral H/He atmosphere)
T_eff = 3562       # Typical Teff for M-giant (T CrB donor is ~M3III)
rho_ph= (9.3*1E-7)     # lower lim
rho_ph2= 8*1E-6 #valor up lim

#es importante mencionar que ambos approach son directamente proporcionales a la densidad
# en la fotosfera, por lo tanto son bastante sensibles a este valor, hay que considerar claro
# que hay valores más coherentes según el principio que macc rate debe ser menor a la mass loss rate
# y coherente con el constraint de la literatura y el de recurrence time 

# --- 2. System Parameters (User Data) ---
P = 227 * 86400     # Period in seconds
M_wd = 1.31 * M_sun # White Dwarf Mass

# Case 1 (Parallax free)
q1 = 0.3198
Msum1 = 1.729 * M_sun
a1 = ((P**2 * G * Msum1) / (4 * (np.pi**2)))**(1/3)
R_rg1 = 54 * R_sun
# Roche Lobe 1
R_rl1 = (a1 * 0.49 * q1**(2/3)) / (0.6 * q1**(2/3) + np.log(1 + q1**(1/3)))

# Case 2 (Gaia parallax)
q2 = 0.4927
Msum2 = 2.06 * M_sun
a2 = ((P**2 * G * Msum2) / (4 * (np.pi**2)))**(1/3)
R_rg2 = 63.4 * R_sun
# Roche Lobe 2
R_rl2 = (a2 * 0.49 * q2**(2/3)) / (0.6 * q2**(2/3) + np.log(1 + q2**(1/3)))


# --- 3. Averaging Values ---
# Note: It is safer to calculate the Donor Mass from the average q and M_wd
# assuming M_wd is fixed, or derive M_donor from M_sum averages.
# Here we follow the strict instruction to average the geometric inputs.

a_avg = (a1 + a2) / 2
R_rg_avg = (R_rg1 + R_rg2) / 2
R_rl_avg = (R_rl1 + R_rl2) / 2
q_avg = (q1 + q2) / 2

# Calculate Donor Mass (M2) using the average mass ratio
# q = M_donor / M_wd  => M_donor = q * M_wd
M_donor = q_avg * M_wd


# --- 4. Thermodynamic Variables (Scale Height) ---

# Let's use specific gas constant R_spec = 8314 / mu (with mu for neutral gas ~ 1.2-1.3)
# Strictly: R_specific = k_B / (mu * m_H) = 8314 / mu_weight
R_specific = 8314.0 / 1.2 # J/kg/K (approx for ionized H is huge, for neutral is smaller) 
# Actually for neutral gas mu ~ 1.2-1.3. 
# c_s = sqrt(k T / mu m_H). 

c_s = np.sqrt((R_gas / 1.2) * T_eff * 1e-3) # Careful with units here. 
# Let's do standard physics: c_s = sqrt(P/rho) = sqrt(k T / mu mH)
k_B = 1.38e-23      #mks
m_H = 1.67e-27      #masa del hidrogeno neutro (es lo mismo que la ma´sa del proton)
c_s = np.sqrt((k_B * T_eff) / (mu * m_H))

# Pressure Scale Height: Hp = Cs^2 / g
# Surface gravity g = G * M_donor / R_rg^2
g = (G * M_donor) / (R_rl_avg**2)
H_p = c_s**2 / g


# --- 5. RLOF Calculation (Ritter Formulation) ---

# F(q) function
# Usually log in these fits is base 10, but in analytic derivations it is often ln. 
# We will use log10 as is common for "log(q)" notation in fitting formulas.
F_q = 1.23 + 0.5 * np.log10(q_avg)

# M_dot_0 Calculation
# Term 1: 2*pi / sqrt(e)
term1 = (2 * np.pi) / np.sqrt(np.e)

# Term 2: (RT/mu)^(3/2) which is exactly c_s^3
term2 = c_s**3

# Term 3: Geometric R_rl^3 / (G * M_donor)
term3 = (R_rl_avg**3) / (G * M_donor)

# Term 4: density and shape factor
term4 = rho_ph * F_q

M_dot_0 = term1 * term2 * term3 * term4

# The Exponential Term
# Delta R = Radius of Star - Radius of Roche Lobe
delta_R = R_rg_avg - R_rl_avg
exp_term = np.exp(delta_R / H_p)

# Final Mass Transfer Rate
M_dot_R = M_dot_0 * exp_term


# --- 6. Output ---
M_dot_R_solar = M_dot_R / (M_sun / (365.25 * 24 * 3600))

print(f"--- T CrB RLOF Estimation ---")
print(f"Average Donor Radius: {R_rg_avg/R_sun:.2f} R_sun")
print(f"Average Roche Lobe:   {R_rl_avg/R_sun:.2f} R_sun")
print(f"Overflow Amount:      {(delta_R/R_sun):.4f} R_sun")
print(f"Scale Height (Hp):    {H_p/R_sun:.4f} R_sun ({H_p:.2e} m)")
print(f"Sound Speed (cs):     {c_s/1000:.2f} km/s")
print(f"-----------------------------")
print(f"M_dot_0 (Base Rate):  {M_dot_0:.2e} kg/s")
print(f"Exponential Factor:   {exp_term:.2e}")
print(f"-----------------------------")
print(f"Final M_dot_RLOF:     {M_dot_R:.3e} kg/s")
print(f"Final M_dot_RLOF:     {M_dot_R_solar:.3e} M_sun/yr")
```

```python
#Now RS Oph OJOOO

# --- 1. Constants (SI Units) ---
G = 6.67e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30    # kg
R_sun = 6.957e8     # m
sigma = 5.67e-8     # Stefan-Boltzmann constant
k_B = 1.38e-23      # Boltzmann constant
m_H = 1.67e-27      # Proton mass

# RS Oph System Data
M_rg = 0.75 * M_sun        # Donor Mass
R_rg = 65 * R_sun          # Donor Radius 
Teff_rg = 3700             # Temperature (Using the hotter estimate)
P = 453.6 * 86400         # Period in seconds
M_wd = 1.35 * M_sun        # White Dwarf Mass
rho_ph = 1e-6              # Photospheric density (approx 10^-9 g/cm^3 for RG)
mu = 1.2                   # Mean molecular weight (Neutral/Molecular envelope)

# --- 4. Calculations ---

# Orbital Separation (Kepler's 3rd Law)
M_total = M_rg + M_wd
a = ((P**2 * G * M_total) / (4 * np.pi**2))**(1/3)

# Mass Ratio q = M_donor / M_accretor
q = M_rg / M_wd

# Roche Lobe Radius (Eggleton)
R_rl = (a * 0.49 * q**(2/3)) / (0.6 * q**(2/3) + np.log(1 + q**(1/3)))

# Surface Gravity (Linear, not log)
# g = GM/R^2
g_surface = (G * M_rg) / (R_rg**2)

# Sound Speed (Isothermal)
# c_s = sqrt(k T / mu m_H)
c_s = np.sqrt((k_B * Teff_rg) / (mu * m_H))

# Pressure Scale Height
H_p = c_s**2 / g_surface

# --- 5. Ritter Mass Transfer Rate ---

# F(q) factor
F_q = 1.23 + 0.5 * np.log10(q) # Using log10 for the fit approximation

# Base Rate (M_dot_0)
# M0 = (2pi / sqrt(e)) * c_s^3 * R_rl^3 / (G M_donor) * rho * F(q)
term_const = (2 * np.pi) / np.sqrt(np.e)
term_cs = c_s**3
term_geom = (R_rl**3) / (G * M_rg)
term_rho = rho_ph * F_q

M_dot_0 = term_const * term_cs * term_geom * term_rho

# Exponential Term
# Delta R = Star Radius - Roche Lobe
delta_R = R_rg - R_rl
exp_term = np.exp(delta_R / H_p)

# Final Rate
M_dot_RLOF = M_dot_0 * exp_term

# Convert to Solar Masses per Year
M_dot_solar = M_dot_RLOF / (M_sun / (365.25 * 24 * 3600))

# --- 6. Output ---
print(f"--- RS Oph RLOF Estimation ---")
print(f"Mass Ratio (q):       {q:.3f}")
print(f"Donor Radius:         {R_rg/R_sun:.2f} R_sun")
print(f"Roche Lobe Radius:    {R_rl/R_sun:.2f} R_sun")
print(f"Overflow (dR):        {(delta_R/R_sun):.2f} R_sun")
print(f"Scale Height (Hp):    {H_p/R_sun:.4f} R_sun")
print(f"------------------------------")
print(f"Base Rate (M_dot_0):  {M_dot_0:.2e} kg/s")
print(f"Exponential Term:     {exp_term:.2e}")
print(f"Final M_dot (RLOF):   {M_dot_solar:.3e} M_sun/yr")
```

Finalizamos con el apporach de Daigne 
Then we look for the RLOF mechanism that is written:
$$\dot M_{RLOF}= \frac{\pi \rho C_s^3}{2k \Omega^2}$$


With $C_s$ and $\rho$ the sound speed and the density at L1, k a factor of order unity and $\Omega$ the angular velocity. For $C_s$ we have a few kilemeters per second, i.e:  5-6 km/s, this asuming that the pressure is the ideal gas pressure in an isothermic case so $\gamma$=1 and $C_s= \gamma kT/\mu m_p$  using Teff from the system we get the value of Cs and we have to look for omega and rho.

```python
import numpy as np

# --- 1. Constants & User Data ---
G = 6.67e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30    # kg
R_sun = 6.957e8     # m
k_B = 1.3806e-23    # Boltzmann constant (J/K)
m_H = 1.6726e-27    # Proton mass (kg)
mu = 1.2            # Mean molecular weight
T_eff = 3562        # K
rho_ph = 1E-7     # kg/m^3
P = 227 * 86400     # s
M_wd = 1.31 * M_sun # kg
k_factor = 1.0      # Factor of order unity

# --- 2. System Parameters (Averaging) ---
# Case 1 (Parallax free)
q1 = 0.3198
Msum1 = 1.729 * M_sun
a1 = ((P**2 * G * Msum1) / (4 * (np.pi**2)))**(1/3)
R_rl1 = (a1 * 0.49 * q1**(2/3)) / (0.6 * q1**(2/3) + np.log(1 + q1**(1/3)))
R_rg1 = 54 * R_sun

# Case 2 (Gaia parallax)
q2 = 0.4927
Msum2 = 2.06 * M_sun
a2 = ((P**2 * G * Msum2) / (4 * (np.pi**2)))**(1/3)
R_rl2 = (a2 * 0.49 * q2**(2/3)) / (0.6 * q2**(2/3) + np.log(1 + q2**(1/3)))
R_rg2 = 63.4 * R_sun

# Averages
a_avg = (a1 + a2) / 2
R_rg_avg = (R_rg1 + R_rg2) / 2
R_rl_avg = (R_rl1 + R_rl2) / 2
q_avg = (q1 + q2) / 2
M_donor = q_avg * M_wd

# --- 3. Daigne Approach Calculation ---

# Angular Velocity (Omega)
Omega = (2 * np.pi) / P

# Sound Speed (Cs) - Isothermal gamma=1
# Cs = sqrt(k_B * T / (mu * m_p))
Cs = np.sqrt((k_B * T_eff) / (mu * m_H))

# Surface Gravity (g) & Scale Height (Hp)
g = (G * M_donor) / (R_rg_avg**2)
Hp = Cs**2 / g

# Density at L1 (rho_L1)
# Using user formula: rho_L1 = (1/sqrt(e)) * rho_ph * exp(-(R_rl - R_star)/Hp)
delta_R = R_rl_avg - R_rg_avg  # Positive if detached, Negative if overflow
term_exp = np.exp(-delta_R / Hp)
rho_L1 = (1 / np.sqrt(np.e)) * rho_ph * term_exp

# Mass Transfer Rate (Daigne)
# M_dot = (pi * rho * Cs^3) / (2 * k * Omega^2)
M_dot_RLOF = (np.pi * rho_L1 * Cs**3) / (2 * k_factor * Omega**2)

# Convert to M_sun/yr
M_dot_solar = M_dot_RLOF / (M_sun / (365.25 * 24 * 3600))

# --- 4. Output ---
print(f"--- T CrB RLOF (Daigne Approach) ---")
print(f"Sound Speed (Cs):     {Cs/1000:.2f} km/s")
print(f"Angular Vel (Omega):  {Omega:.2e} rad/s")
print(f"Scale Height (Hp):    {Hp/R_sun:.4f} R_sun")
print(f"Radius Gap (R_rl-R*): {(delta_R/R_sun):.4f} R_sun")
print(f"Density at L1:        {rho_L1:.3e} kg/m^3")
print(f"------------------------------------")
print(f"Mass Transfer Rate:   {M_dot_RLOF:.3e} kg/s")
print(f"Mass Transfer Rate:   {M_dot_solar:.3e} M_sun/yr")
```

```python
#RS Oph

import numpy as np

# --- 1. Constants ---
G = 6.67e-11        # m^3 kg^-1 s^-2
M_sun = 1.989e30    # kg
R_sun = 6.957e8     # m
k_B = 1.3806e-23    # Boltzmann constant (J/K)
m_H = 1.6726e-27    # Proton mass (kg)
mu = 1.2            # Mean molecular weight
k_factor = 1.0      # Factor of order unity

# --- 2. RS Oph System Data ---
P = 453.6 * 86400   # Period in seconds
M_wd = 1.35 * M_sun # White Dwarf Mass
M_rg = 0.75 * M_sun # Red Giant Mass
R_rg = 65 * R_sun   # Red Giant Radius
T_eff = 3700        # Teff (K)
rho_ph = 1E-7     # Photospheric density (Assumed similar to T CrB/Typical RG)

# --- 3. Derived Parameters ---

# Orbital Separation (a)
M_total = M_wd + M_rg
a = ((P**2 * G * M_total) / (4 * np.pi**2))**(1/3)

# Mass Ratio (q)
q = M_rg / M_wd

# Roche Lobe Radius (R_rl) - Eggleton
R_rl = (a * 0.49 * q**(2/3)) / (0.6 * q**(2/3) + np.log(1 + q**(1/3)))

# Angular Velocity (Omega)
Omega = (2 * np.pi) / P

# Sound Speed (Cs) - Isothermal gamma=1
Cs = np.sqrt((k_B * T_eff) / (mu * m_H))

# Surface Gravity (g) & Scale Height (Hp)
g = (G * M_rg) / (R_rg**2)
Hp = Cs**2 / g

# --- 4. Daigne Calculation ---

# Density at L1 (rho_L1)
# rho_L1 = (1/sqrt(e)) * rho_ph * exp(-(R_rl - R_*)/Hp)
delta_R = R_rl - R_rg  # Positive means Detached (R_rl > R_rg)
term_exp = np.exp(-delta_R / Hp)
rho_L1 = (1 / np.sqrt(np.e)) * rho_ph * term_exp

# Mass Transfer Rate (Daigne)
# M_dot = (pi * rho * Cs^3) / (2 * k * Omega^2)
M_dot_RLOF = (np.pi * rho_L1 * Cs**3) / (2 * k_factor * Omega**2)

# Convert to M_sun/yr
M_dot_solar = M_dot_RLOF / (M_sun / (365.25 * 24 * 3600))

# --- 5. Output ---
print(f"--- RS Oph RLOF (Daigne Approach) ---")
print(f"Roche Lobe Radius:    {R_rl/R_sun:.2f} R_sun")
print(f"Red Giant Radius:     {R_rg/R_sun:.2f} R_sun")
print(f"Radius Gap (R_rl-R*): {(delta_R/R_sun):.2f} R_sun")
print(f"Scale Height (Hp):    {Hp/R_sun:.4f} R_sun")
print(f"------------------------------------")
print(f"Density at L1:        {rho_L1:.3e} kg/m^3")
print(f"Mass Transfer Rate:   {M_dot_solar:.3e} M_sun/yr")
```
